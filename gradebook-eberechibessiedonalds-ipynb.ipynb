```python
#STEP ONE? WE IMPORT PANDAS AND CREATE THE STORAGE #Pandas helps us display data in a clean table format.

import pandas as pd

#this serves as a database that stores all students record.

students = []

# THIS TRACKS THE ID NUMBER FOR EACH NEW STUDENT. IT STARTS FROM 1 AND ASCENDS...1,2,3..

next_id = 1
```

```python
#STEP TWO? GRADE CALCULATOR FUNCTION #This function takes a score and returns the grade.
#We'll use this anytime we need to assign a grade.

def calculate_grade(average):
    if average >= 70:
        return "A"
    elif average >= 60:
        return "B"
    elif average >= 50:
        return "C"
    elif average >= 40:
        return "D"
    else:
        return "F"
```

```python
#STEP THREE? ADD STUDENT_BRACKET FUNCTION #THIS FUNCTION ADDS A NEW STUDENT TO THE SYSTEM

def add_student(): 
    #WE WILL USE GLOBAL SO THAT THE FUNCTION CAN ACCESS AND UPDATE NEXT_ID FROM OUTSIDE THE FUNCTION.

    global next_id, students

  #input bracket asks the user to type somethng
  #whatever they type gets stored in the variable.

    name = input("Enter student_name:" )

  #int() converts the typed text into a number because input() always gives us text by default

    test1 = int(input("Enter Test 1 score (0-100): "))
    test2 = int(input("Enter Test 2 score (0-100): "))

#CALCULATE AVERAGE AUTOMATICALLY

    average = (test1 + test2)/ 2

#ASSIGN GRADE USING OUR FUNCTION FROM STEP 2

    grade = calculate_grade(average)

#CREATE ONE STUDENT RECORD AS A DICTIONARY

    student = {
      "ID"      : next_id,
      "Name"    : name,
      "Test1"   : test1,
      "Test2"   : test2,
      "Average" : average,
      "grade"   : grade
  }

  #ADD THIS STUDENT TO OUR STUDENTS LIST

    students.append(student)

  #MOVE ID FORWARD FOR THE NEXT STUDENT

    next_id += 1

  #CONFIRM TO THE USER THAT IT WORKED

    print(f"\nStudent '{name}' was added successfully with Grade {grade}!")


  
```

```python
#STEP FOUR? VIEW_STUDENT() FUNCTION  #THIS FUNCTION DISPLAYS ALL STUDENTS ON A NEAT TABLE

def view_students():
    
    #CHECK IF THERE ARE ANY STUDENTS FIRST
    if len(students) == 0:
        print("\n No students found")
        return

     #CONVERT OUR LIST INTO A PANDAS DATAFRAME- LIKE A SPREADSHEET TABLE

df = pd.DataFrame(students)

     #DISPLAY THE TABLE NEATLY

print("\n ALL STUDENTS")
print(df.to_string(index=False))
```

```python
#STEP FIVE? UPDATE STUDENT()  FUNCTION   #THIS FUNCTION UPDATES A STUDENT'S EXISTING INFORMATION

def update_student():
    global students

  #ASK THE USER WHICH STUDENT HE WANTS TO UPDATE THEIR INFORMATION, USING THE STUDENT ID

    student_id = int(input("Enter student ID to update:"))

  #SEARCH THROUGH ALL THE STUDENTS TO FIND THE RIGHT ONE THE USER WANTS
  #WE USE A VARIABLE TO TRACK IF WE FIND THE RIGHT ONE

    found = False
    for student in students:
        if student ["ID"] == student_id:
            found = True

           #ASK USER WHAT THEY WANT TO UPDATE

        print("\nWhat would you like to update?")
        print("1. Name")
        print("2. Test 1 Score")
        print("3. Test 2 Score")

        choice = input("Enter choice (1/2/3): ")


        #UPDATE BASED ON USER'S CHOICE

        if choice == "1":
            student["Name"] = input("Enter new name: ")
            
        elif choice == "2":
            student["Test1"] = int(input("Enter new Test 1 Score: "))

        elif choice == "3":
            student["Test2"] = int(input("Enter new Test 2 Score: "))

        else:
            print("Invalid choice!")
            return

    #RECALCULATE AVERAGE AND GRADE AFTER UPDATE

        student["Average"] = (student["Test1"] + student["Test2"]) / 2
        student["Grade"] = calculate_grade(student["Average"])

     #SHOW THE UPDATED STUDENT INFO

        print(f"\nStudent updated successfully!")
        print(f"Name: {student['Name']}")
        print(f"Test1: {student['Test1']}")
        print(f"Test2: {student['Test2']}")
        print(f"Average: {student['Average']}")
        print(f"Grade: {student['Grade']}")

            #IF THE STUDENT WAS NOT FOUND IN THE SYSTEM, LET THE USER KNOW

        if not found:
         print(f"\nStudent with ID {student_id} not found")
```

```python
#STEP SIX? DELETE_STUDENT() FUNCTION #THIS FUNCTION HELPS TO REMOVES A STUDENT FROM THE SYSTEM

def delete_student():
    global students

    #ASK THE USER WHICH STUDENT TO DELETE

    student_id = int(input("Enter Student ID to delete: "))

    #SEARCH FOR THE STUDENT IN THE LISTS OF STUDENTS

    found = False
    for student in students:
        if student["ID"] == student_id:
            found = True
            name = student["Name"]

            #REMOVE THE STUDENT FROM THE LIST

            students.remove(student)
            print(f"\nStudent '{name}' has been deleted successfully!")
            break    #BREAK MEANS THE SYSTEM STOPS SEARCHING ONCE THE STUDENT IS FOUND.

    if not found:
        print(f"\nStudent not found!")
```
